```sql
SELECT 
    '2015' AS year_sales,
    COUNT(*) AS num_of_sales
FROM 
    dsci_504.orders
WHERE 
    EXTRACT(YEAR FROM ord_date) = 2015;
```

```sql
SELECT 
    TO_CHAR(ord_date, 'YYYY-MM') AS month,
    COUNT(*) AS num_of_sales
FROM 
    dsci_504.orders
WHERE 
    EXTRACT(YEAR FROM ord_date) = 2015
GROUP BY 
    TO_CHAR(ord_date, 'YYYY-MM')
ORDER BY 
    month;
```

```sql
SELECT 
    EXTRACT(YEAR FROM ord_date) AS year,
    COUNT(*) AS num_of_sales
FROM 
    dsci_504.orders
GROUP BY 
    EXTRACT(YEAR FROM ord_date)
ORDER BY 
    year;
```

```sql
SELECT 
    TO_CHAR(cus_join_date, 'YYYY-MM') AS month,
    COUNT(*) AS num_customers
FROM 
    dsci_504.customers
WHERE 
    EXTRACT(YEAR FROM cus_join_date) = 2018
GROUP BY 
    TO_CHAR(cus_join_date, 'YYYY-MM')
ORDER BY 
    month;
```

```sql
SELECT 
    EXTRACT(YEAR FROM cus_join_date) AS year,
    COUNT(*) AS num_customers
FROM 
    dsci_504.customers
GROUP BY 
    EXTRACT(YEAR FROM cus_join_date)
ORDER BY 
    year;
```

```sql
SELECT 
    EXTRACT(YEAR FROM cus_join_date) AS year,
    COUNT(*) AS num_customers
FROM 
    dsci_504.customers
GROUP BY 
    EXTRACT(YEAR FROM cus_join_date)
ORDER BY 
    num_customers DESC
LIMIT 1;
```

```sql
SELECT 
    c.cus_id,
    o.ord_id,
    c.cus_join_date,
    o.ord_date,
    o.ord_ship_date,
    o.ord_ship_date - c.cus_join_date AS days_between
FROM 
    dsci_504.customers c
JOIN 
    dsci_504.orders o ON c.cus_id = o.cus_id
WHERE 
    EXTRACT(YEAR FROM o.ord_date) = 2018;
```

```sql
SELECT 
    MIN(o.ord_ship_date - c.cus_join_date) AS shortest_delay,
    MAX(o.ord_ship_date - c.cus_join_date) AS longest_delay
FROM 
    dsci_504.customers c
JOIN 
    dsci_504.orders o ON c.cus_id = o.cus_id
WHERE 
    EXTRACT(YEAR FROM o.ord_date) = 2016;
```

```sql
CREATE TABLE public.shimano_recall_customers (
    cus_id INT,
    cus_last_name VARCHAR,
    cus_first_name VARCHAR,
    cus_state VARCHAR,
    cus_phone VARCHAR
);
```

```sql
INSERT INTO public.shimano_recall_customers (cus_id, cus_last_name, cus_first_name, cus_state, cus_phone)
SELECT 
    c.cus_id, 
    c.cus_last_name, 
    c.cus_first_name, 
    c.cus_state, 
    c.cus_phone
FROM 
    dsci_504.customers c
JOIN 
    dsci_504.orders o ON c.cus_id = o.cus_id
JOIN 
    dsci_504.products p ON o.prod_id = p.prod_id
WHERE 
    p.prod_name LIKE '%Shimano XTR%';
```

```sql
WITH build_cte (id) AS ( SELECT build_id 
FROM dsci_504.builds 
WHERE comp_brake IN (SELECT comp_id::varchar from components 
WHERE comp_name LIKE '%XTR%') OR comp_shifter IN (SELECT comp_id::VARCHAR from components 
WHERE comp_name LIKE '%XTR%') OR comp_cassette IN (SELECT comp_id::VARCHAR from components 
WHERE comp_name LIKE '%XTR%') OR comp_chain_ring IN (SELECT comp_id::VARCHAR from components 
WHERE comp_name LIKE '%XTR%') OR comp_derail IN (SELECT comp_id::VARCHAR from components 
WHERE comp_name LIKE '%XTR%') ) , ord_cte (ord_id,cus_id,build_id) AS ( SELECT ord_id ,cus_id ,build_id 
FROM orders ) , cus_cte (id,last_name,first_name,state, phone) AS ( SELECT cus_id ,cus_last_name ,cus_first_name ,cus_state ,cus_phone 
FROM customers ) SELECT cus_cte.id ,cus_cte.last_name ,cus_cte.first_name ,cus_cte.state ,cus_cte.phone 
FROM cus_cte JOIN ord_cte ON ord_cte.cus_id = cus_cte.id JOIN build_cte ON build_cte.id = ord_cte.build_id
```

```sql
SELECT 
    p.prod_manufacturer,
    p.prod_name,
    p.prod_description,
    p.country_origin,
    SUM(o.order_tot) AS total_order_total
FROM 
    dsci_504.products p
JOIN 
    dsci_504.orders o ON p.prod_id = o.prod_id
GROUP BY 
    p.prod_manufacturer, 
    p.prod_name, 
    p.prod_description, 
    p.country_origin
ORDER BY 
    total_order_total DESC, 
    p.prod_manufacturer, 
    p.prod_name;
```

```sql
SELECT 
    o.ord_id,
    c.cus_id,
    p.prod_name,
    AVG(o.order_tot) OVER() AS avg_order_total,
    SUM(o.order_tot) OVER() AS sum_all_orders,
    p.prod_price - o.order_tot AS price_order_difference,
    o.order_tot,
    p.prod_price
FROM 
    dsci_504.orders o
JOIN 
    dsci_504.products p ON o.prod_id = p.prod_id
JOIN 
    dsci_504.customers c ON o.cus_id = c.cus_id
ORDER BY 
    o.ord_id;
```

```sql
SELECT 
    c.cus_id,
    c.cus_state,
    c.cus_phone,  -- If considering phone as non-PII; otherwise exclude this
    o.order_return_amount AS return_amount,
    o.order_return_date AS return_date
FROM 
    dsci_504.customers c
JOIN 
    dsci_504.orders o ON c.cus_id = o.cus_id
WHERE 
    o.order_status = 'Returned'
ORDER BY 
    o.order_return_date;
```
